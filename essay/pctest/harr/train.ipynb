```python
import cv2
```

```python
model = cv2.CascadeClassifier('5di/cascade.xml')
```

```python
img = cv2.imread('tennis/t0.jpg')
```

```python
aa =  cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
tennis = model.detectMultiScale(aa, 1.3, 500)
```

```python
tennis
```

```python
for (x,y,w,h) in tennis:
        # img：要画的图片，(x,y)：起始坐标 ，(x+w,y+h)：宽高，颜色，线条宽度
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
import matplotlib.pyplot as plt
```
